```python
import pandas as pd

df = pd.read_csv("gdp_co2_by_country.csv")
df.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["Population"], df["CO2"])
plt.xlabel("Population")
plt.ylabel("CO2 Emissions")
plt.title("Population vs CO2 Emissions")
plt.show()
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["GDP Per Capita"], df["Per Capita CO2"])
plt.xlabel("GDP Per Capita")
plt.ylabel("CO2 Emissions per Capita")
plt.title("GDP vs CO2 Emissions")
plt.show()
```

```python
df_clean = df[df["Year"] == 2020]
```

```python
import matplotlib.pyplot as plt

plt.scatter(df_clean["GDP Per Capita"], df_clean["Per Capita CO2"])
plt.xlabel("GDP Per Capita")
plt.ylabel("CO2 Emissions per Capita")
plt.title("GDP vs CO2 Emissions (2020)")
plt.show()
```
